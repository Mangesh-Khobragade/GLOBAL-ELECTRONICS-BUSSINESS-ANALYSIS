```python
mergedf2 =mergedf2.drop_duplicates()
```

```python
mergedf2.dtypes
```

```python
# Removing dollar signs and commas
mergedf2['Unit Price USD'] = mergedf2['Unit Price USD'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
mergedf2['Unit Cost USD'] = mergedf2['Unit Cost USD'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
mergedf2['Profit'] = (mergedf2['Unit Price USD'] - mergedf2['Unit Cost USD']) * mergedf2['Quantity']
```

```python
mergedf2['Birthday'] = pd.to_datetime(mergedf2['Birthday'])
```

```python
mergedf2['Order Date'] = pd.to_datetime(mergedf2['Order Date'])
mergedf2['Delivery Date'] = pd.to_datetime(mergedf2['Delivery Date'])

# Calculate delivery time delta for non-missing rows
mergedf2['Delivery Delta'] = (mergedf2['Delivery Date'] - mergedf2['Order Date']).dt.days

# Calculate mean delivery time (ignoring NaN values)
mean_delivery_time = mergedf2['Delivery Delta'].mean()

# Fill missing delivery dates based on mean delivery time
mergedf2['Delivery Date'] = mergedf2['Delivery Date'].fillna(mergedf2['Order Date'] + pd.Timedelta(days=mean_delivery_time))

# At this point, 'Delivery Date' is still in datetime format

# Optionally drop the 'Delivery Delta' column if no longer needed
mergedf2.drop(columns=['Delivery Delta'], inplace=True)

# Format the 'Delivery Date' to mm/dd/yyyy as a string
mergedf2['Delivery Date'] = mergedf2['Delivery Date'].dt.strftime('%m/%d/%Y')

# Display the updated DataFrame
print('successful')
```
